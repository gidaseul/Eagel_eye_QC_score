```python
import pandas as pd
import ast
import re
from rapidfuzz import process as rf_process, fuzz

# 1. CSV 로드
url = "/Users/10fingers/Desktop/Eagle_eye_gds/final_result/Naver/Naver_kakao_merged(5192).csv"
df = pd.read_csv(url, encoding="utf-8-sig")

# 2. score_mapping 정의
score_mapping = {
    5: ["뷔페", "갈비 무한 리필", "고기 무한 리필", "초밥 무한 리필", "참치 무한 리필", "일본 가정식", "오무라이스 전문점", "회전 초밥", "연어 요리", "연어 덮밥", "간장 새우덮밥", "스테이크/와규 덮밥", "BBQ 플래터 (바베큐)", "대게 요리 레스토랑", "개인 돌판/화로구이", "파인다이닝 레스토랑", "샤브샤브", "일본 나베 요리", "시카고피자", "스페인 레스토랑", "규카츠", "곱창/막창", "오마카세", "차이니즈/딤섬", "랍스터 전문점", "텐동(튀김덮밥)", "스테이크 전문점", "수제 햄버거 전문점"],
    4: ["인도 레스토랑", "미국 가정식", "멕시코 음식점", "프랑스 코스 요리", "학센", "일본식 카레", "일본 라멘", "아부라소바/비벼먹는 라멘", "함바그", "코스 양식 요리", "화덕 피자 전문점", "피자 레스토랑", "파스타 레스토랑", "초밥 전문점", "닭갈비 전문점", "조개 구이", "양고기 구이", "돼지고기 구이", "소고기 구이", "갈비찜", "양꼬치", "철판 요리", "훠궈", "월남쌈", "불고기", "해물탕 / 해물뚝배기", "마라탕", "이색치킨"],
    3: ["부대찌개", "치킨", "경양식 돈까스", "닭볶음탕", "닭발", "오리고기 구이", "메밀 전문점", "짜글이", "삼계탕", "족발", "찜닭", "태국 레스토랑", "베트남 레스토랑", "해물찜", "육회 전문점", "즉석떡볶이", "냉동/대패 삼겹살", "곱도리탕", "중식, 차이니즈", "생선구이", "낙곱새", "순대 요리 전문점", "두부요리", "우육탕면"],
    2: ["퓨전 떡볶이", "술집겸 밥집", "일본식 우동점", "한우 뚝배기 불고기", "한상차림식 한식백반"],
    1: ["동네 중국집", "분식", "카프레제", "수프전문점"]
}
all_keywords = [item for sublist in score_mapping.values() for item in sublist]
pass_list = ["카페", "디저트", "베이커리"]

# 3. 유사도 보조 함수
def split_by_korean_token(text):
    text = re.sub(r"[^가-힣a-zA-Z0-9]", " ", text)
    return text.strip().split()

def token_similarity(input_keyword, target_keyword):
    input_tokens = split_by_korean_token(input_keyword)
    target_tokens = split_by_korean_token(target_keyword)
    if not input_tokens or not target_tokens:
        return 0.0
    scores = [max([fuzz.ratio(i, t) for t in target_tokens]) for i in input_tokens]
    return sum(scores) / len(scores)

def score_from_target(target):
    for score, items in score_mapping.items():
        if target in items:
            return score
    return 0

# 4. 메인 매칭 함수 (Theme, Category, Name을 바탕으로 유사도 판단)
def smart_combined_match(theme_str, category_str, name_str, threshold=70, token_threshold=75):
    try:
        theme_list = ast.literal_eval(theme_str) if isinstance(theme_str, str) else []
        category_list = [c.strip() for c in category_str.split(',')] if isinstance(category_str, str) else []
    except Exception:
        return None, None, None, "❌ 파싱 실패"

    # (0) 제외 카테고리 (카페, 디저트 관련 카테고리는 매칭에서 제외하도록) -> 이건 나중에 그냥 입력 전에 처리하도록 
    if any(ban_word in cat for ban_word in pass_list for cat in category_list):
        return None, None, None, "🚫 카페/디저트/베이커리 category → 매칭 제외"

    # (1순위) score_mapping에 category가 직접 존재 (category에 있는 값이 score_mapping 정확히 일치할 때)
    for cat in category_list:
        for score, items in score_mapping.items():
            if cat in items:
                return cat, score, 100.0, f"category '{cat}' → score_mapping 직접 매칭 (점수: {score})"

    # (2순위) score_mapping의 항목과 category가 부분 일치 (category에 있는 값이 score_mapping 부분적으로 일치할 때 ex) category = 중식당, target = 중식, 차이니즈)
    for cat in category_list:
        for score, items in score_mapping.items():
            for target in items:
                if cat in target or target in cat:
                    return target, score, 100.0, f"category '{cat}' ⊂ '{target}' → score_mapping 부분 매칭 (점수: {score})"

    # 후보 저장 (여러가지 상황에서 가장 유사도 높은 것을 선택하도록)
    candidates = []

    # (1-1) theme_topic 없음 fallback → category + name 기반 (부분적으로 크롤링 할 때 Theme 관련 키워드가 없을 수도 있음.)
    if not theme_list:
        fallback_sources = category_list + [name_str]
        for source in fallback_sources:
            match, similarity, _ = rf_process.extractOne(source, all_keywords)
            if similarity >= threshold:
                score = score_from_target(match)
                candidates.append(("fallback", match, score, similarity, f"🏳️ fallback 매칭 '{match}' ({similarity:.1f})"))
        # 끝까지 fallback도 없으면 종료
        if not candidates:
            return None, None, None, "❌ theme_topic 없음 + fallback 실패"

    # (1-2) category <-> theme 포함 매칭 (theme_topic의 키워드 중 category와 관련 있는 것이 있으면 그것부터 매칭)
    for keyword in theme_list:
        for cat in category_list:
            if keyword in cat or cat in keyword:
                score = score_from_target(keyword)
                if score > 0:
                    candidates.append(("category-theme 포함", keyword, score, 100.0, f"✅ category '{cat}' ⊃ theme '{keyword}' (점수: {score})"))

    # (1-3) 카테고리-테마 교차 키워드 기반 유사도 평가
    for cat in category_list:
        for keyword in theme_list:
            sim = fuzz.partial_ratio(cat, keyword)
            if sim >= threshold:
                for score, items in score_mapping.items():
                    for target in items:
                        if keyword in target or target in keyword:
                            candidates.append(("교차 유사도", target, score, sim, f"🔁 교차 유사도 '{cat}' <-> '{keyword}' → '{target}'"))

    # (1-4) score_mapping 직접 포함 매칭 (테마 키워드가 score_mapping 항목 중 하나와 직접 일치하거나 포함되면 매칭)
    # 키워드 전체 loop 중 "가장 관련성 높은 매칭 후보"를 선택
    for keyword in theme_list:
        for score, items in score_mapping.items():
            for target in items:
                if target in keyword or keyword in target:
                    candidates.append(("직접 포함", target, score, 100.0, f"✅ '{keyword}' ⊂ '{target}' → 직접 포함"))

    # (1-5) 유사도 기반 (가장 유사한 score_mapping 키워드를 찾고, 기준(70 이상)이면 매칭 인정)
    for keyword in theme_list:
        match, similarity, _ = rf_process.extractOne(keyword, all_keywords)
        if similarity >= threshold:
            score = score_from_target(match)
            candidates.append(("유사도 기반", match, score, similarity, f"🔍 유사도 매칭 '{keyword}' → '{match}' ({similarity:.1f})"))

    # (1-6) 토큰 기반 (3번이 실패했을 경우에 토큰 단위로 나뉘어서 유사도 확인)
    for keyword in theme_list:
        for target in all_keywords:
            sim = token_similarity(keyword, target)
            if sim >= token_threshold:
                score = score_from_target(target)
                candidates.append(("토큰 유사도", target, score, sim, f"🧪 토큰 유사도 '{keyword}' → '{target}' ({sim:.1f})"))
    # 최종 후보 중 최고 선택
    if candidates:
        best = sorted(candidates, key=lambda x: (x[2], x[3]), reverse=True)[0]  # score > similarity 기준
        return best[1], best[2], best[3], f"[{best[0]}] {best[4]}"
    
    return None, None, None, "❌ 모든 매칭 실패"

# 5. 적용
df[["matched_theme_name", "matched_theme_score", "matched_similarity", "match_log"]] = df.apply(
    lambda row: pd.Series(smart_combined_match(row["theme_topic"], row["category"], row["name"])),
    axis=1
)

# 6. 결과 출력
print(df[["naver_id", "name", "category", "theme_topic", "matched_theme_name", "matched_theme_score", "matched_similarity", "match_log"]].head(30).to_string(index=False))
```

```python
import pandas as pd
import ast
import re
from rapidfuzz import fuzz
from rapidfuzz import process as rf_process

# 1. CSV 로드
url = "/Users/10fingers/Desktop/Eagle_eye_gds/final_result/Naver/Naver_kakao_merged(5192).csv"
df = pd.read_csv(url, encoding="utf-8-sig")
df = df[~df["category"].str.contains("카페|디저트|베이커리", na=False)]

# 2. score_mapping 정의
score_mapping = {
    5: ["뷔페", "갈비 무한 리필", "고기 무한 리필", "초밥 무한 리필", "참치 무한 리필", "일본 가정식", "오무라이스 전문점", "회전 초밥", "연어 요리", "연어 덮밥", "간장 새우덮밥", "스테이크,와규 덮밥", "BBQ 플래터 (바베큐)", "대게 요리 레스토랑", "개인 돌판,화로구이", "파인다이닝 레스토랑", "샤브샤브", "일본 나베 요리", "시카고피자", "스페인 레스토랑", "규카츠", "곱창,막창", "오마카세", "차이니즈,딤섬", "랍스터 전문점", "텐동(튀김덮밥)", "스테이크 전문점", "수제 햄버거"],
    4: ["인도 레스토랑", "미국 가정식", "멕시코 음식점", "프랑스 코스 요리", "학센", "일본식 카레", "일본 라멘", "아부라소바/비벼먹는 라멘", "함바그", "코스 양식 요리", "화덕 피자 전문점", "피자 레스토랑", "파스타 레스토랑", "초밥 전문점", "닭갈비 전문점", "조개 구이", "양고기 구이", "돼지고기 구이", "소고기 구이", "갈비찜", "양꼬치", "철판 요리", "훠궈", "월남쌈", "불고기", "해물탕,해물뚝배기", "마라탕", "이색치킨"],
    3: ["부대찌개", "치킨", "경양식 돈까스", "닭볶음탕", "닭발", "오리고기 구이", "메밀 전문점", "짜글이", "삼계탕", "족발", "찜닭", "태국 레스토랑", "베트남 레스토랑", "해물찜", "육회 전문점", "즉석떡볶이", "냉동,대패 삼겹살", "곱도리탕", "중식, 차이니즈", "생선구이", "낙곱새", "순대 요리 전문점", "두부요리", "우육탕면"],
    2: ["퓨전 떡볶이", "술집겸 밥집", "일본식 우동점", "한우 뚝배기 불고기", "한상차림식 한식백반"],
    1: ["동네 중국집", "분식", "카프레제", "수프전문점"]
}
all_keywords = [kw for items in score_mapping.values() for kw in items]

# 3. 필터용 리스트
pass_list     = ["카페","디저트","베이커리","아이스크림","제과점"]
generic_cats  = ["양식","한식","중식","일식","분식"]

# 4. 토큰 분리
def split_by_korean_token(text):
    text = re.sub(r"[^가-힣a-zA-Z0-9]", " ", str(text))
    return text.strip().split()

# 5. 타니모토 계수 (빈도 수 까지 확인하기 위해서 사용하는 방식)
def tanimoto_coefficient(a: str, b: str) -> float:
    set_a = set(split_by_korean_token(a))
    set_b = set(split_by_korean_token(b))
    if not set_a and not set_b:
        return 1.0
    inter = set_a & set_b
    union = set_a | set_b
    return len(inter) / len(union)

# 6. base_score 추출
def score_from_target(target):
    for sc, items in score_mapping.items():
        if target in items:
            return sc
    return 0

# 7. 매칭 함수
def smart_match(theme_str, category_str, name_str,
                sim_threshold=60, name_thresh=70, theme_thresh=70):
    # 파싱
    try:
        theme_list    = ast.literal_eval(theme_str) if isinstance(theme_str, str) else []
    except:
        theme_list    = [theme_str] if theme_str else []
    category_list = [c.strip() for c in str(category_str).split(",")]

    # 1) 카테고리 패스 필터
    if any(b in cat for b in pass_list for cat in category_list):
        return pd.Series([None, None, None, None, 0, "🚫 패스 카테고리", []],
                         index=output_cols)

    has_theme = bool(theme_list)
    combined_src = " ".join(theme_list + category_list + [name_str])

    candidates = []
    # 가중치 비율에 대해서
    WEIGHTS = {"cat":0.25, "theme":0.25, "name":0.2, "mix":0.2, "tanimoto":0.1}

    for target in all_keywords:
        base_score = score_from_target(target)

        # 2) mention_count 계산
        #   theme_flags *2
        theme_flags = 0
        for th in theme_list:
            if th and (th == target or th in target or target in th):
                theme_flags += 1
            elif fuzz.partial_ratio(th, target) >= theme_thresh:
                theme_flags += 1

        #   name_flag
        name_sim  = fuzz.partial_ratio(name_str, target)
        name_flag = 1 if name_sim >= name_thresh else 0

        #   cat_flag (generic category 무시)
        cat_flag = 0
        for cat in category_list:
            if cat and (cat == target or cat in target or target in cat):
                if not(has_theme and cat in generic_cats):
                    cat_flag = 1
                    break

        mention_count = theme_flags*2 + name_flag + cat_flag*0.5

        # 3) 유사도 피처
        cat_sim   = max((fuzz.partial_ratio(cat, target) for cat in category_list), default=0)
        theme_sim = max((fuzz.partial_ratio(th, target) for th in theme_list), default=0)
        mix_sim   = fuzz.partial_ratio(combined_src, target)
        tanimoto_sim = tanimoto_coefficient(combined_src, target)*100

        # 4) combined_sim & final_score
        combined_sim = (
            WEIGHTS["cat"]   * cat_sim +
            WEIGHTS["theme"] * theme_sim +
            WEIGHTS["name"]  * name_sim +
            WEIGHTS["mix"]   * mix_sim +
            WEIGHTS["tanimoto"] * tanimoto_sim
        )
        final_score = base_score * (combined_sim/100)

        candidates.append({
            "target":       target,
            "base_score":   base_score,
            "mention_count":mention_count,
            "combined_sim": combined_sim,
            "final_score":  final_score
        })

    # 5) 정렬·선택
    candidates.sort(key=lambda x:(x["mention_count"], x["final_score"]), reverse=True)
    best   = candidates[0]
    passed = (best["mention_count"] >= 2
              and best["combined_sim"] >= sim_threshold
              and best["base_score"]    > 0)

    log = (f"대상:'{best['target']}', base:{best['base_score']}, "
           f"mention:{best['mention_count']}, sim:{best['combined_sim']:.1f}, "
           f"final:{best['final_score']:.1f}, "
           f"{'통과' if passed else '임계치 미달'}")

    failed = [
        {"target":c["target"],
         "mention":c["mention_count"],
         "final": round(c["final_score"],1)}
        for c in candidates[1:] if c["mention_count"]>0
    ][:3]

    return pd.Series([
        best["target"]   if passed else None,
        best["base_score"] if passed else None,
        best["combined_sim"] if passed else None,
        best["final_score"]   if passed else None,
        best["mention_count"],
        log,
        failed
    ], index=output_cols)

# 8. 컬럼 이름 정의 후 적용
output_cols = [
    "matched_theme_name","matched_theme_score","matched_similarity",
    "matched_final_score","mention_count","match_log","failed_candidates"
]
df_new = df.apply(lambda r: smart_match(
    r["theme_topic"], r["category"], r["name"]
), axis=1)
df = pd.concat([df, df_new], axis=1)

# 9. 결과 확인
print(df[[
    "naver_id","name","category","theme_topic",
    *output_cols
]].head(30).to_string(index=False))
```

```python
import pandas as pd
from shapely.geometry import Point
from shapely import wkt

# 🔹 polygon dict 로딩 함수
def load_polygons_from_df(df, name_col="location", polygon_col="polygon_str"):
    polygons = {}
    for _, row in df.iterrows():
        try:
            name = row[name_col]
            poly = wkt.loads(row[polygon_col])
            polygons[name] = poly
        except:
            continue
    return polygons

# 🔹 점수 계산 함수
def calculate_location_score(row, hotplace_polys, campus_polys, new_hot_keywords):
    lat, lng = row.get("gps_latitude"), row.get("gps_longitude")
    address = str(row.get("address", ""))
    distance = pd.to_numeric(row.get("distance_from_subway", None), errors="coerce")

    if pd.isna(lat) or pd.isna(lng):
        return pd.Series({
            "location_score": 0.0,
            "score_source": None,
            "score_flag": False,
            "score_reason": "missing_coordinates"
        })

    point = Point(lng, lat)
    region_type = "서울" if "서울" in address else "수도권"

    for name, poly in hotplace_polys.items():
        if point.within(poly):
            return pd.Series({
                "location_score": 5.0,
                "score_source": "hotplace",
                "score_flag": True,
                "score_reason": "ok"
            })

    for keyword in new_hot_keywords:
        if keyword in address:
            return pd.Series({
                "location_score": 4.0,
                "score_source": "new_hotplace",
                "score_flag": True,
                "score_reason": "ok"
            })

    for _, poly in campus_polys.items():
        if point.within(poly):
            return pd.Series({
                "location_score": 4.0,
                "score_source": "campus",
                "score_flag": True,
                "score_reason": "ok"
            })

    if pd.isna(distance):
        return pd.Series({
            "location_score": 0.0,
            "score_source": None,
            "score_flag": False,
            "score_reason": "missing_distance"
        })

    if distance <= 900:
        return pd.Series({
            "location_score": 3.0,
            "score_source": "distance",
            "score_flag": True,
            "score_reason": "ok"
        })

    return pd.Series({
        "location_score": 0.0,
        "score_source": "distance",
        "score_flag": False,
        "score_reason": "unmatched"
    })
```

```python
# 🔹 데이터 로드
df_store = pd.read_csv("/Users/10fingers/Desktop/Eagle_eye_gds/final_result/Naver/Naver_kakao_merged(5192).csv")
df_hotplace = pd.read_csv("seoul_hotspots_wkt.csv")
df_campus = pd.read_csv("campus_polygons.csv")

# 🔹 polygon dict 생성
hotplace_polys = load_polygons_from_df(df_hotplace, "location", "polygon_str")
campus_polys = load_polygons_from_df(df_campus, "campus_name", "polygon_str")

# 🔹 신규 핫플 키워드 정의
new_hot_keywords = ["삼성역", "코엑스", "익선동", "샤로수길", "송리단길", "해방촌", "후암동", "서촌"]

# 🔹 점수 계산 적용
df_scored = df_store.apply(
    lambda row: calculate_location_score(row, hotplace_polys, campus_polys, new_hot_keywords),
    axis=1
)

# 🔹 결과 병합
df_result = pd.concat([df_store, df_scored], axis=1)

# # 🔹 저장
# df_result.to_csv("scored_store_data.csv", index=False, encoding="utf-8-sig")
print(df_result[["naver_id", "name", "category", "address","distance_from_subway","gps_latitude","gps_longitude","location_score", "score_source", "score_flag", "score_reason"]].head(30).to_string(index=False))
```

```python
import pandas as pd
import json

# 1. 엑셀 파일 로드
file_path = "네이버 카테고리 기반 목록표  2.xlsx"
xls = pd.ExcelFile(file_path)
df_raw = xls.parse("네이버 카테고리 목록표 정리")

# 2. 각 세트 컬럼 정의
cols_0 = ["대분류", "중분류", "소분류", "목록"]
cols_1 = ["대분류.1", "중분류.1", "소분류.1", "목록.1"]
cols_2 = ["대분류.2", "중분류.2", "소분류.2", "목록.2"]

# 3. 세트별 정리 및 병합
dfs = []
for cols in [cols_0, cols_1, cols_2]:
    sub_df = df_raw[cols].copy()
    sub_df.columns = ["대분류", "중분류", "소분류", "목록"]
    sub_df = sub_df.dropna(subset=["대분류", "중분류", "목록"], how="any")
    dfs.append(sub_df)

df_combined = pd.concat(dfs, ignore_index=True)

# 4. 문자열 목록을 리스트로 변환
def parse_list(cell):
    if isinstance(cell, str):
        return [item.strip('- ').strip() for item in cell.strip().split("\n") if item.strip()]
    return []

df_combined["목록"] = df_combined["목록"].apply(parse_list)

# 5. JSON 저장
category_json = df_combined.to_dict(orient="records")

with open("category_mapping.json", "w", encoding="utf-8") as f:
    json.dump(category_json, f, ensure_ascii=False, indent=2)

print("✅ category_mapping.json 저장 완료")
```
